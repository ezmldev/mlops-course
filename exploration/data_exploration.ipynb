In this notebook we are taking a look at our dataframe and visualize it.

First, import `course_helpers` and load the dataset:

```python
from course_helpers import load_dataset

df = load_dataset()
```

## Data Visualization

Let's see the most common words in the data.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


def draw_wordcloud(df):
    all_text = " ".join(df["content"])

    wordcloud = WordCloud(
        width=800, height=400, background_color="white", random_state=42
    ).generate(all_text)

    plt.figure(figsize=(5, 3))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```

```python
draw_wordcloud(df)
```

How about the most common words in those rows that aren't classfied as spam?

```python
draw_wordcloud(df[df["is_spam"] == 0])
```

How about the most common words in those rows that aren classfied as spam?

```python
draw_wordcloud(df[df["is_spam"] == 1])
```

We can see right away that different words stand out for spam and non-spam messages. This is a strong indication that a machine learning model can be trained to distinguish between the two.
